# Q1. Display different categories of food having highest and lowest varieties and plot it graphically

```python
df["Category"].value_counts()
```

```python
plt.figure(figsize=(14,10))
plt.title("Food Categories")
sns.countplot(x="Category",data=df)
plt.xlabel("categories")
plt.ylabel("count")
plt.grid()
plt.show()
```

# Inference



From the above graph it can be concluded that

1) The highest sales is of Tea & Coffee

2) The lowest sales is of Salads & Desserts

# Q2. Create a pie chart that shows proportions of each cateogry of food 

```python
plt.figure(figsize=(8,9))
plt.title("Category %")
df["Category"].value_counts().plot(kind="pie",autopct="% 1.2f%%",explode=(0.09,0,0,0,0,0,0,0.09,0.09))
plt.show()
```

# Q3. Display Category of food having maximum amount of cholesterol and plot graph for same.

```python
max_cholestrol=df["Cholesterol"].max()
df[df["Cholesterol"]==max_cholestrol]
```

# Q4. Draw a graph Cholesterol Vs Food Categories

```python
plt.figure(figsize=(10,8))
df.pivot_table('Cholesterol','Category').plot(kind='bar')
plt.show()
```

# Q5. Display top 10 Smoothies & Shakes having high calories ?

```python
ss=df[df["Category"] == "Smoothies & Shakes"]
ss[["Item", "Calories"]].sort_values("Calories",ascending=False).head(10)
```

# Q6. What is the mean (average) calorie count of each category?

```python
cal_grp=df.groupby(by="Category")
cal_grp["Calories"].mean()
```

```python
plt.figure(figsize=(7,6))
plt.title("Calorie Count")
cal_grp["Calories"].mean().plot(kind="bar",color="cyan")
plt.ylabel("Average")
plt.show()
```

# Q8. Display top 5 Item which contributes maximum amount of Proteins in a diet.
